This notebook contains basic TensorFlow examples.



```python
import tensorflow as tf
import numpy as np

print("TF:", tf.__version__)

scalar = tf.constant(7)
vector = tf.constant([10, 20, 30])
matrix = tf.constant([[1., 2.], [3., 4.]])

print(scalar.shape, vector.shape, matrix.shape)
```

This code shows how TensorFlow represents numbers.
Scalars, vectors, and matrices are different types of tensors.
The shape tells how the data is structured.


```python
A = tf.constant([[10., 7.], [3., 4.]])
print(A + 10)
print(A * 2)
```
